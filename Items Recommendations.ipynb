# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

# Data Preprocessing

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv',header = None)
dataset
```

```python
transaction = []
for i in range(0,7501):
   transaction.append([str(dataset.values[i,j]) for j in range(0,20)])
       
```

# trainning the apriori model on the dataset

```python
!pip install apyori
```

```python
from apyori import apriori
rules = apriori(transactions = transaction,min_support = 0.003,min_confidence = 0.3,min_lift = 3,min_length = 3,max_length = 3)
```

# Visualization

```python
results = list(rules)
print(results)
```

## For better Visualization

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    support = [result[1] for result in results]
    confidence = [result[2][0][2] for result in results]
    lift = [result[2][0][3] for result in results]
    return list(zip(lhs,rhs,support,confidence,lift))
```

```python
df = pd.DataFrame(inspect(results),columns=['Left Hand Side','Right hand Side','Support','Confidence','Lift'])
df
```

```python
df = df.nlargest(n=45,columns='Lift')#sorting the df by lift column in descending order
df.reset_index(inplace=True)
df
```

```python
df = df.iloc[:,1:]
```
